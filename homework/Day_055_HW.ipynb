- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

np.random.seed(5)
%matplotlib inline
```

```python
estimator = [("kmeans_5", KMeans(n_clusters=5)),
            ("kmeans_10", KMeans(n_clusters=10)),
             ("kmeans_bad_init", KMeans(n_clusters=10, n_init=1,
                                                        init="random"))]
```

```python
fignum = 1
titles = ["5_cluster", "10_cluster","10_cluster_bad_int"]
for name, est in estimator:
    fig = plt.figure(fignum, figsize=(4,3))
    ax = Axes3D(fig, rect = [0, 0, .95,1],elev= 100, azim = 90)
    #fit data
    est.fit(X)
    labels = est.labels_
    
    ax.scatter(X[:,2],X[:,0],X[:,3],
              c= labels.astype(np.float), edgecolor = 'k')
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum -1])
    ax.dist=12
    fignum = fignum+1
#plot the grounf truth
fig = plt.figure(fignum, figsize=(4,3))
ax = Axes3D(fig, rect=[0,0,.95,1],elev =40, azim = 90)

index = [("cls0",0),("cls1",1),("cls2",2)]
for name, label in index:
    ax.text3D(X[y == label,2].mean(),   #??? # ax.text3D : 將不同label的資料標上個別物種類名稱，裡面 X[y == label, 3].mean() 用在調整 text 的 X Y Z 軸位置
              horizontalalignment='center',
              X[y == label,0].mean(),
              X[y == label,3].mean() +2,name,
              horizontalalignment='center',
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1,2,0]).astype(np.float)                          ##用途??
ax.scatter(X[:,2], X[:,0],X[:,3], c= y, edgecolor ='k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title("ground_truth")
ax.dist = 8
```
